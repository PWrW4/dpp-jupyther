```python
def oblicz(danet,dane,danep,danerh):
    zdt = []
    for i in range(0,len(danet)):
        tc = float(danet[i]) - 273.15
        liczba =6.112 * math.exp((17.67 *tc)/(tc + 243.5))

        z= 0.002277*(danep[i]+((1255/float(dane1t[i]))+0.05)*liczba*float(danerh[i])/100)
        zdt.append(z)

    return zdt
```
